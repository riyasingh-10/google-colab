```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,]
```

```python
plt.hist(dataset)
```

```python

## Z Score
outliers=[]

def detect_outliers(data):
  threshold=3 ## 3 std deviation
  mean=np.mean(data)
  std=np.std(data)

  for i in data:
    z_score=(i-mean)/std
    if np.abs(z_score)> threshold:
      outliers.append(i)

  return outliers
```

```python
detect_outliers(dataset)
```
