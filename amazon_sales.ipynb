```python
df = pd.read_csv("amazon.csv")
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Clean and convert 'discounted_price' and 'actual_price'
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)

# Clean and convert 'discount_percentage'
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False).astype(float)

# Clean and convert 'rating'
# The rating column might have some non-numeric values (e.g., '4.4 out of 5 stars', or just 'nan' as a string).
# Use errors='coerce' to turn unparseable values into NaN.
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Clean and convert 'rating_count'
# The rating_count column contains commas and might have non-numeric entries.
# Use errors='coerce' to turn unparseable values into NaN.
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')


print("Data types after conversion:")
display(df.info())
```

```python
display(df.describe())
```

```python
# Calculate the number of missing values per column
missing_values_count = df.isnull().sum()

# Calculate the percentage of missing values per column
missing_values_percentage = (df.isnull().sum() / len(df)) * 100

# Create a DataFrame to display the results
missing_values_summary = pd.DataFrame({
    'Missing Count': missing_values_count,
    'Missing Percentage': missing_values_percentage
})

# Display only columns with missing values, sorted by percentage
display(missing_values_summary[missing_values_summary['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False))
```

```python
missing_values = df.isnull().sum()
missing_percentage = df.isnull().sum() / len(df) * 100

missing_data = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percentage
})
display(missing_data[missing_data['Missing Count']>0].sort_values(by = 'Missing Percentage'))
```

```python
print("dataframe before dropping the values", df.shape)

df.dropna(subset = ['rating' , 'rating_count'], inplace = True)

print("dataframe after dropping the values" , df.shape)

#verify the values
missing_values = df.isnull().sum()
missing_percentage = df.isnull().sum() / len(df) * 100
missing_data = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percentage
})
display(missing_data[missing_data['Missing Count']>0].sort_values(by = 'Missing Percentage'))
```

```python
high_discounted_products = df[df['discount_percentage']>50]
display(high_discounted_products)
```

```python
plt.figure(figsize=(12,10))
sns.histplot(data =high_discounted_products, x='discounted_price',bins = 50 , kde = True)
plt.title("Distribution of Discounted Prices for High Discount Products")
plt.xlabel("Discounted Price")
plt.ylabel("Frequency\\\n")
plt.tight_layout()
plt.show()
```

```python
# Sort high_discounted_products by discounted_price and select the top 10 for better readability
top_10_discounted_products = high_discounted_products.sort_values(by='discounted_price', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_discounted_products, x='discounted_price', y='product_name')
plt.title('Top 10 Products by Discounted Price (High Discount Products)')
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```
